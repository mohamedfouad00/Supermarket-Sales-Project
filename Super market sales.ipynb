IQR

<M>

```python
def detect_outliers_iqr(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    upper_bond=q3+1.5*iqr
    lower_bond=q1-1.5*iqr
    return df[(df[column]>upper_bond) | (df[column]<lower_bond)]
    
```

```python
def iqr_outliers_values(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    upper_bond=q3+1.5*iqr
    lower_bond=q1-1.5*iqr
    return upper_bond,lower_bond
```

```python
upper_bond,lower_bond=iqr_outliers_values(df,'Tax 5%')
print(f'Upper bond = {upper_bond}\nLower bond = {lower_bond}')
```

```python
detect_outliers_iqr(df,'Tax 5%')
```

```python
upper_bond,lower_bond=iqr_outliers_values(df,'Gross income')
print(f'Upper bond = {upper_bond}\nLower bond = {lower_bond}')
```

```python
detect_outliers_iqr(df,'Gross income')
```

### 1.3.2 Formatting and adding columns

<m>

date format

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
try:
    pd.to_datetime(df["Date"])
    print("Valid datetime format")
except ValueError:
    print("Invalid datetime format")
```

```python
df["Date"]=df["Date"].dt.date
```

rating status

```python
df["rating_status"]="undefined"
```

```python
for index,row in df.iterrows():
    if row["Rating"]<=6:
        df.loc[index,"rating_status"]="low rating"
    elif row["Rating"]<=8:
        df.loc[index,"rating_status"]="medium rating"
    elif row["Rating"]>8:
        df.loc[index,"rating_status"]="high rating"    
```

Order Staus

```python
df["order_status"]="undifiend"
```

```python
for index,row in df.iterrows():
    if row["Quantity_sold"]<=3:
        df.loc[index,"order_status"]="small order"
    elif row["Quantity_sold"]<=6:
        df.loc[index,"order_status"]="medium order"    
    elif row["Quantity_sold"]>6:
        df.loc[index,"order_status"]="large order"    
```
